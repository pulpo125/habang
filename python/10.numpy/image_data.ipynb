```python
im_color = cv2.imread('data/apple_img.jpg', cv2.IMREAD_COLOR) # BGR
im_color = cv2.cvtColor(im_color, cv2.COLOR_BGR2RGB) # BGR -> RGB 로 변환
```

```python
for i in range(0,40):
    im_color[120+i, 300+i:340-i, :] = [45, 88, 170]
plt.imshow(im_color)
```

```python
for i in range(0,70):
    im_color[120+i, 300+i:540-i, :] = [45, 88, 170]
    im_color[120-i, 300+i:540-i, :] = [45, 88, 170]
plt.imshow(im_color)
```

```python
im_color[150:160, 250:350, :] = [55,84,164] # RGB
im_color[200:210, 250:350, :] = [55,84,164]
im_color[150:210, 250:260, :] = [77,164,55]
im_color[150:210, 340:350, :] = [77,164,55]
plt.imshow(im_color)
```
