```python
plt.subplots(2,3)
# fig, axes = plt.subplots(2,3) => fig 에는 그림이 들어가고 axes 에는 축 array 가 들어간다. (튜플로 리턴)
```

```python
# (2, 3) Subplots 그리기
plt.rcParams['figure.figsize'] = (8, 4)
data = np.arange(1, 51)
fig, axes = plt.subplots(2,3)

axes[0, 0].plot(data)
axes[0, 1].plot(data ** 2)
axes[0, 2].plot(data ** 3)
axes[1, 0].plot(data % 10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data // 20)

plt.tight_layout()
plt.show()
```

- Scatter plot

```python
import numpy as np

plt.rcParams['figure.figsize'] = (5, 4)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 250

plt.scatter(x, y, s=area, c=colors) # s 파라미터에 각 점 크기를 반영할 수 있다
plt.show
```

- boxplot 이산형 변수

```python
x = ['Chemistry', 'History', 'Biology', 'Music', 'Spanish'] # 과목
y = [90, 75, 85, 70, 95] # 점수

plt.xlabel('Subject')
plt.ylabel('Score')
plt.bar(x, y, alpha=0.2, color='r') # alpha:투명도
```

```python
plt.xlabel('Score')
plt.ylabel('Subject')
plt.barh(x, y, alpha=0.2, color='r')
```

```python
plt.rcParams['figure.figsize'] = (8, 4)

x_label = ['Chemistry', 'History', 'Biology', 'Music', 'Spanish'] # 과목
x = np.arange(len(x_label)) # 각 과목의 숫자 위치 [0, 1, 2, 3, 4]

y_1 = [90, 75, 85, 70, 95] # Emily
y_2 = [85, 90, 70, 60, 80] # Benjamin

# 넓이 
width = 0.35

fig, axes = plt.subplots()
axes.bar(x - width/2, y_1, width, alpha=0.5)
axes.bar(x + width/2, y_2, width, alpha=0.8)
plt.xticks(x)
axes.set_xticklabels(x_label) # 눈금 설정
plt.legend(['Emily', 'Benjamin'], loc=0) # loc 위치 default: 0 = best
plt.xlabel('Subject')
plt.ylabel('Score')

plt.show()
```

- areaplot (선 아래 채우는 시각화)

```python
x = np.arange(1,21)
y = np.random.randint(5,10,20)
plt.fill_between(x, y, color='green', alpha=0.3)
plt.plot(x, y, color='green', alpha=0.8)
plt.show()
```

```python
x = np.arange(1, 10, 0.05)
y_1 =  np.cos(x)+1
y_2 =  np.sin(x)+1
y_3 = y_1 * y_2 / np.pi
plt.fill_between(x, y_1, color="green", alpha=0.1)
plt.fill_between(x, y_2, color="blue", alpha=0.2)
plt.fill_between(x, y_3, color="red", alpha=0.3)
```

- boxplot

```python
plt.rcParams['figure.figsize'] = (5, 4)

spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100

data = np.concatenate([spread, center, flier_high, flier_low])
plt.boxplot(data)
```

```python
plt.boxplot(data, vert=False) # 가로
```

```python
outlier_marker = {'markerfacecolor' : 'r', 'marker' : 'D'}
plt.boxplot(data, flierprops = outlier_marker)
plt.show()
```
