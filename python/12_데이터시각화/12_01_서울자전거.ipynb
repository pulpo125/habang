### 데이터시각화 matplotlib 서울자전거

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../0_data/SeoulBikeData.csv')
df.head(2)
```

- 온도에 대한 시각화

```python
plt.figure(figsize=(5,4))
plt.hist(df['Temperature'], bins=range(-20,41,5))
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.xticks(range(-20,41,5))
plt.show()
```

- boxplot

```python
plt.boxplot(df['Temperature'])
plt.title('Temperature')
plt.show()
```

```python
plt.hist(df['Rented Bike Count'], bins=range(0,3601,300))
plt.xlabel('Rented Bike Count')
plt.ylabel('Frequency')
plt.xticks(range(0,3601,300))
```

```python
plt.boxplot(df['Rented Bike Count'])
plt.title('Rented Bike Count')
```

- 계절별 자전거 대여량 (평균)

```python
season_bike_series = df.groupby('Seasons')['Rented Bike Count'].mean()
season_bike_series
```

```python
# 평균 자전거 대여량: 봄, 여름, 가을, 겨울 순으로 추출
seasons = ['Spring', 'Summer', 'Autumn', 'Winter']
mean_bike_cnt = [season_bike_series[season] for season in seasons]
# mean_bike_cnt = season_bike_series[seasons]
plt.bar(seasons, mean_bike_cnt)
```

- 기온과 자전거 대여량 관계

```python
plt.scatter(df['Temperature'], df['Rented Bike Count'])
plt.xlabel('Temperature')
plt.ylabel('Rented Bike Count')
```
