```python
df.dropna(how='all') # 모든 값이 null인 경우만 해당 행을 drop 한다
```

- missing_data_test

```python
df = pd.read_csv('data/missing_data_test.csv', encoding='cp949', index_col = '연도')
```

```python
df.drop(index=[2019])
```

```python
# 특정 컬럼 제거
# df.drop(['제품3, 제품4'], axis=1)
df.drop(columns=['제품3', '제품4'])
```

```python
df.drop(index=[2019], columns=['제품3', '제품4'])
```

```python
# 행 기준, 한 개라도 null이 있으면 제거
df.dropna(how = 'any') # default: axis = 0
```

```python
# 특정 컬럼에 결측치가 있는 행만 제거
# 제품1 컬럼에 널값이 있으면 행을 제거 다른 행은 널값 있어도 됌
df.dropna(subset=['제품1'])
```

```python
df = pd.read_csv('data/missing_data_test.csv', encoding='cp949', index_col = '연도')
```

```python
# 특정 행에 결측치가 있는 열만 제거
df.dropna(axis=1, subset=[2015])
```

```python
df.dropna(axis=1, subset=[2016, 2019])
```

- 널 값 채우기

```python
df.fillna(0)
```

```python
# bfill: 결측치가 있는 열의 그 다음 값으로 채움
# 그 다음 값이 없는 경우는 계속 Null
df.fillna(method='bfill')
```

```python
# ffill: 결측치가 있는 열의 그 이전 값으로 채움
# 그 이전 값이 없는 경우는 계속 Null
df.fillna(method='ffill')
```

```python
# 특정 열에 특정 값으로 널을 채움
values = {'제품1': 100, '제품3': 300,'제품4':400}
df.fillna(value=values)
```

```python
values = {'제품1': df['제품1'].mean(), '제품2': df['제품2'].mean(), '제품3':df['제품3'].mean(),'제품4':df['제품4'].mean()}
df.fillna(value=values)
```
