### 판다스 Merge와 중복제거 실습

```python
import pandas as pd
import numpy as np
```

- merge 실습

```python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})

df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})

data = {'Col1': ['a', 'a', 'b', 'a', 'b'],
        'Col2': [1, 1, 1, 2, 2],
        'Col3': [1, 1, 2, 2, 2]}
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # 조던이랑 우즈가 없어짐 -> 디폴트가 inner join 이기 때문이다.
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
df1 = df1.rename(columns={'고객명':'이름'})
pd.merge(df1, df2, left_on='이름', right_on='고객명')
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명', how='outer')
```
