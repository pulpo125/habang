```python
kto_201901['기준년월'] = '2019-01'
```

```python
kto_201901['국적'].unique()
```

```python
# 대륙 목록
continents_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
len(continents_list)
```

```python
# 국적에 대륙이 포함된 행을 제거
condition = ~kto_201901['국적'].isin(continents_list)
kto_201901_country = kto_201901.loc[condition, :]
kto_201901_country.shape
```

```python
kto_201901_country['국적'].unique()
```

```python
kto_201901_country.reset_index(drop=True, inplace=True) # drop=True 인덱스 버림
```

```python
kto_201901_country.tail()
```

```python
# 대륙에 몇 개의 나라가 속해있는지
continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 \
+ ['아프리카']*2 + ['기타대륙'] + ['교포']
len(continents)
```

```python
kto_201901_country['대륙'] = continents
kto_201901_country.head()
```

```python
kto_201901_country.shape
```

```python
# 관광객 비율
kto_201901_country['관광객비율(%)'] = round(kto_201901_country['관광'] / kto_201901_country['계'] * 100, 1)
kto_201901_country.head()
```

```python
# 관광객비율이 높은 나라 순으로 정렬(내림차순)
kto_201901_country.sort_values('관광객비율(%)', ascending=False).head(10)
```

```python
# 관광객비율이 낮은 나라 순으로 정렬(오름차순)
kto_201901_country.sort_values('관광객비율(%)', ascending=True).head(10)
```

```python
# 대륙별 관광객 비율 평균
kto_201901_country.groupby('대륙')[['관광객비율(%)']].mean().sort_values('관광객비율(%)')
```

```python
kto_201901_country.pivot_table(index='대륙', values='관광객비율(%)').sort_values('관광객비율(%)')
```

```python
kto_201901_country.pivot_table(index='대륙', values='관광객비율(%)', aggfunc=['mean', 'count'])
```

```python
# 중국 국적인 행만 추출
kto_201901_country.loc[kto_201901_country['국적'] == '중국', :]
```

```python
# 2019년 1월 우리나라에 방문한 관광 객의 총 숫자
tourlist_sum = kto_201901_country['관광'].sum()
```

```python
# 관광객비율_국가(%)
kto_201901_country['관광객비율_국가(%)'] = round(kto_201901_country['관광'] / tourlist_sum * 100, 1)
kto_201901_country.head()
```
