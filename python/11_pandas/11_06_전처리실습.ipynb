- 여러 파일 연결

```python
from glob import glob
glob('data/kto_files/*')
```

```python
# 2020년 (kto_201001.xlsx)
kto_201001 = pd.read_excel('data/kto_files/kto_201001.xlsx', header=1, usecols='A:G', skipfooter=4)
kto_201001.head()
```

```python
def create_kto_data(yyyy, mm):
    # 데이터 불러오기
    file_path = f'data/kto_files/kto_{yyyy}{mm}.xlsx'
    df = pd.read_excel(file_path, header=1, usecols='A:G', skipfooter=4)

    # 전처리
    # 1. 기준년월 컬럼 추가
    df['기준년월'] = f'{yyyy}-{mm}'

    # 2. 국적 컬럼에서 대륙 제거하고 국가만 남기기
    continents_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']
    condition = ~df['국적'].isin(continents_list)
    df = df.loc[condition, :]
    df.reset_index(drop=True, inplace=True)

    # 3. 대륙 컬럼 추가
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']
    df['대륙'] = continents 

    # 4. 관광객비율 컬럼 추가
    df['관광객비율(%)'] = round(df['관광'] / df['계'] * 100, 1)

    # 5. 관광객비율_국가 컬럼 추가
    tourlist_sum = df['관광'].sum()
    df['관광객비율_국가(%)'] = round(df['관광'] / tourlist_sum * 100, 1)

    return df
```

```python
# 전체 데이터(120개)를 연결할 데이터프레임 초기화
df_total = pd.DataFrame()

# 201001~ 201912
for yyyy in range(2010, 2020):
    for mm in range(1, 13):
        df_tmp = create_kto_data(yyyy, str(mm).zfill(2))
        df_total = pd.concat([df_total, df_tmp], ignore_index=True) # 누적

df_total.shape
```
