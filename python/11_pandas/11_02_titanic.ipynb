- 문제 1

```python
# age가 null, 컬럼: ['pclass', 'age', 'who', 'fare', 'survived']
# sort: fare 기준으로 내림차순

condition = (df['age'].isnull())
df.loc[condition, ['pclass', 'age', 'who', 'fare', 'survived']].sort_values(by='fare', ascending=False)
```

```python
# deck과 age가 모두 null이 아니어야 함
# 컬럼: ['deck', 'pclass', 'age', 'who', 'fare', 'survived']
# deck을 기준으로 오름차순, 이후 age를 기준으로 오름차순
df.loc[(df['deck'].notnull()) & (df['age'].notnull()), ['deck', 'pclass', 'age', 'who', 'fare', 'survived']].sort_values(by=['deck', 'age'])
```

```python
# who의 값이 child 이면서 alone의 값도 True인 경우
# 컬럼: ['survived', 'pclass', 'age', 'who', 'alone']
# 정렬: age, 내림차순
df.loc[(df['who'] == 'child') & (df['alone'] == True), ['survived', 'pclass', 'age', 'who', 'alone']].sort_values(by='age', ascending=False)
```

```python
# who가 남자 or 여자, age 는 null 이 아님

# df.loc[(df['who'] == 'man') | (df['who'] == 'woman') & (df['age'].notnull()), ['who', 'age']].sort_values(by=['who', 'age'], ascending=[True, False])
df.loc[(df['who'].isin(['man', 'woman'])) & (df['age'].notnull()), ['who', 'age']].sort_values(by=['who', 'age'], ascending=[True, False])
```
