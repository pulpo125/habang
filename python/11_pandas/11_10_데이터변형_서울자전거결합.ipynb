```python
import pandas as pd
```

```python
df_2018 = pd.read_csv('data/서울특별시 공공자전거 대여이력 정보_2018년_1분기1.csv', encoding='cp949')
df_2018.head()
```

```python
# df_2018['대여일시'] = df_2018['대여일시'].astype('datetime64')
df_2018['대여일시'] = pd.to_datetime(df_2018['대여일시']) # 효율적
df_2018['반납일시'] = pd.to_datetime(df_2018['반납일시'])
```

```python
df_2018['대여연도'] = df_2018['대여일시'].dt.year
df_2018['대여월'] = df_2018['대여일시'].dt.month
df_2018['대여일'] = df_2018['대여일시'].dt.day
```

```python
df_2018.head(2)
```

- 날씨와 서울시 자전거

```python
df = pd.read_csv('../0_data/SeoulBikeData.csv', encoding='utf-8')
df.head()
```

```python
df.shape
```

### 목표: 두 가지 데이터를 결합하여 비교하기
- 1. 날짜별 자전거 대여 개수 비교
- 2. 온도와 평균 자전거 대여 시간 

```python
# Date 컬럼의 자료형을 datetime으로 변경
# pd.to_datetime(df['Date']) => 디폴트로 MM/DD/YYYY로 전환
# 그러므로, dayfirst = True 를 설정하여 DD/MM/YYYY로 설정해준다.
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
df['Date'].head()
```

```python
# 대여연도, 대여월, 대여일 컬럼 생성
df['대여연도'] = df['Date'].dt.year
df['대여월'] = df['Date'].dt.month
df['대여일'] = df['Date'].dt.day
df.head(2)
```

```python
# 대여월, 대여일별 자전거대여수 합계 구하기
df.pivot_table(index=['대여월', '대여일'], values='Rented Bike Count', aggfunc='sum')
#  df.groupby(['대여월','대여일'])[['Rented Bike Count']].sum()
```

- 두 데이터 Merge ( 서울시 건별 자전거 데이터, 서울시 날씨 자전거 데이터)

```python
# month_day_rent_df: 서울시 날씨 자전거 데이터
month_day_rent_df = df.groupby(['대여월','대여일'])[['Rented Bike Count']].sum()
month_day_rent_df.rename(columns={'Rented Bike Count':'자전거대여량(날씨)'}, inplace=True)
month_day_rent_df
```

```python
# month_day_rent_df2018 (서울시 건별 자전거 데이터)
month_day_rent_df2018 = pd.DataFrame(df_2018.groupby(['대여월','대여일']).size(), columns=['자전거대여량(건별)'])
month_day_rent_df2018 # (90,1)
```

```python
# index를 기준으로 병합(Join 함수)
# 왼쪽: 날씨 서울시 자전거, 오른쪽: 건별 서울시 자전거
month_day_rent_df.join(month_day_rent_df2018)
```

```python
inner_join_df = month_day_rent_df.join(month_day_rent_df2018, how='inner')
inner_join_df
```

```python
inner_join_df.describe()
```

```python
inner_join_df.corr() # 높은 양의 상관관계
```

```python
# Temperature - Snowfall 시간별 각 날씨 평균 집계
# inner_join_weather_df
```

```python
month_day_renttime_df2018 = df_2018.groupby(['대여월','대여일'])[['이용시간(분)']].mean()
month_day_renttime_df2018
```

```python
month_day_wheater_df = df.groupby(['대여월', '대여일'])[['Temperature','Wind speed (m/s)', 'Rainfall(mm)', 'Snowfall (cm)']].mean()
month_day_wheater_df
```

```python
inner_join_weather_renttime = month_day_wheater_df.join(month_day_renttime_df2018, how='inner')
inner_join_weather_renttime
```
