```python
import numpy as np
import matplotlib.pyplot as plt

# 활성화, XOR
```

- 활성화 함수

> Sigmoid

```python
x = np.arange(-6, 6, 0.1)
x[:10]
```

```python
x[-10:]
```

```python
# sigmoid 함수
def sigmoid(x):
    sigmoid =  1 / (1 + np.exp(-x))
    return sigmoid

y = [sigmoid(val) for val in x]
```

```python
plt.plot(x,y)
plt.title('sigmoid')
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

y = [tanh(val) for val in x]
```

```python
plt.plot(x,y)
plt.title('tanh')
```

```python
# ReLU 함수
def ReLU(x):
    if x > 0:
        y = x
    else:
        y = 0
    return y

y = [ReLU(val) for val in x]

# def ReLU(x):
#     return x if x > 0 else 0
```

```python
plt.plot(x,y)
plt.title('ReLU')
```

- XOR 시각화

```python
xor_list = [(0,0), (0,1), (1,0), (1,1)]
```

```python
for i in xor_list:
    plt.scatter(i[0], i[1])
```

```python
# 강사님
x, y = zip(*xor_list) # 반대로 빼줌
plt.scatter(x, y)
plt.title('XOR')
plt.scatter(x, y, c=['blue','red','red','blue'])
```
